# Градиентный спуск для минимизации линейной регрессии
Будем рассматривать случай парной регресси, то есть моделью будет являться $a(x) = w_1 \cdot x + w_0$, где x - это единственный признак.

Для начала сгенерируем случайную выборку.

```python
import torch

count = 10

x = torch.rand(count, 2)

print(x)
```

В качестве функции потерь будем использовать функционал усредненной квадратичной разности.

```python
from torch.autograd import Variable

w = Variable(torch.ones(1, 2), requires_grad=True)

Q = 0
for i in range(count):
    Q += (w[0][1] * x[i][0] + w[0][0] - x[i][1]) ** 2
Q *= (1 / count)

k = 0.2

Q.backward(retain_graph=True)

print(w.grad)
```

```python
import matplotlib.pyplot as plt

x_axis = []
y_axis = []

for i in range(count):
    x_axis += [x[i][0]]
    y_axis += [x[i][1]]
plt.plot(x_axis, y_axis, 'ro')
plt.axis([-1, 1, -1, 1])

plt.plot([-1, 1], [w[0][1] * -1 + w[0][0], w[0][1] * 1 + w[0][0]])
plt.show()
```

```python
w.data -= w.grad * 0.2
Q.backward(retain_graph=True)
```

```python
x_axis = []
y_axis = []

for i in range(count):
    x_axis += [x[i][0]]
    y_axis += [x[i][1]]
plt.plot(x_axis, y_axis, 'ro')
plt.axis([-1, 1, -1, 1])

plt.plot([-1, 1], [w[0][1] * -1 + w[0][0], w[0][1] * 1 + w[0][0]])
plt.show()
```

```python
for i in range(100):
    w.data -= w.grad * 0.0001
    Q.backward(retain_graph=True)
```

```python
x_axis = []
y_axis = []

for i in range(count):
    x_axis += [x[i][0]]
    y_axis += [x[i][1]]
plt.plot(x_axis, y_axis, 'ro')
plt.axis([-1, 1, -1, 1])

plt.plot([-1, 1], [w[0][1] * -1 + w[0][0], w[0][1] * 1 + w[0][0]])
plt.show()
```
